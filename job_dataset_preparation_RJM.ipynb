```python
import pandas as pd
df = pd.read_csv("monster_com-job_sample.csv")
print("Dataset shape:", df.shape)
df.head()
```

```python
df = df[["job_title", "organization", "location", "job_description"]]
df = df.dropna(subset=["job_title", "organization", "location", "job_description"])
print("After dropping nulls:", df.shape)
```

```python
df.head()
```

```python
df = df[df["job_description"].str.len() > 100].copy()
df.reset_index(drop=True, inplace=True)
print("After filtering by length:", df.shape)
df.head(5)
```

```python
df.to_csv("jobs_cleaned_day1.csv", index=False)
print("Saved cleaned CSV with", len(df), "rows.")
```
